```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

while True:
    text = input("\nEnter paragraph to summarize:\n")

    max_len = int(input("Enter max summary length: "))
    min_len = int(input("Enter min summary length: "))

    summary = summarizer(text, max_length=max_len, min_length=min_len, do_sample=False, early_stopping=True)

    print("\n--- SUMMARY ---")
    print(summary[0]['summary_text'])

    again = input("\nDo you want to summarize another text? (yes/no): ").strip().lower()
    if again not in ['yes', 'y']:
        print("Exiting summarizer. Goodbye!")
        break
```
