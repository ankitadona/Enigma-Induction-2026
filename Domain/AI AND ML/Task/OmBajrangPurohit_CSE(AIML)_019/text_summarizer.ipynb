## Welcome

### Import libraries

```python
from transformers import pipeline
```

### Import pre-trained model

```python
model_name = "t5-small"
```

### Summarizer code block

```python
print("Loading model... (this may take a minute the first time)")
summarizer = pipeline("summarization", model=model_name, device = 0)
```

### Summary Generator

```python
text = input("Enter the text here: \n")
word_count = len(text.split())

max_l = int(word_count / 2) 
min_l = int(word_count / 4)

if max_l < 10:
    max_l = 20
    min_l = 5
    
print("\nGenerating Summary...")
summary = summarizer(
        text, 
        max_length=max_l, 
        min_length=min_l, 
        do_sample=False
    )
print("\nSummery\n")
print(summary[0]['summary_text'])
```
